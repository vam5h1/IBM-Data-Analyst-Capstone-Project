### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
import pandas as pd

# Step 1: Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Identify missing values for all columns
missing_values = df.isnull().sum()

# Step 3: Display missing values (only columns with missing data)
print("🔍 Missing Values by Column:\n")
print(missing_values[missing_values > 0].sort_values(ascending=False))
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Set figure size and plot heatmap
plt.figure(figsize=(18, 10))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')

# Step 3: Display the plot
plt.title("Heatmap of Missing Values", fontsize=16)
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.tight_layout()
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
import pandas as pd

# Step 1: Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Count missing rows in the 'Employment' column
missing_employment = df['Employment'].isnull().sum()

# Step 3: Display the result
print(f"Number of missing rows in 'Employment' column: {missing_employment}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
import pandas as pd

# Step 1: Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Find the most frequent value in the 'Employment' column
most_common_employment = df['Employment'].value_counts().idxmax()

# Step 3: Display the result
print(f"The most frequent value in the 'Employment' column is: {most_common_employment}")
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
import pandas as pd

# Step 1: Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Find the most frequent value in 'Employment'
most_common_employment = df['Employment'].value_counts().idxmax()

# Step 3: Fill missing values with the most frequent value
df['Employment'].fillna(most_common_employment, inplace=True)

# Step 4: Verify that there are no more missing values
missing_after = df['Employment'].isnull().sum()
print(f"Missing values in 'Employment' column after imputation: {missing_after}")
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Impute missing values in 'Employment' with the most frequent value
most_common_employment = df['Employment'].value_counts().idxmax()
df['Employment'].fillna(most_common_employment, inplace=True)

# Step 3: Plot the distribution
plt.figure(figsize=(10,6))
df['Employment'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Employment Status After Imputation', fontsize=14)
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
