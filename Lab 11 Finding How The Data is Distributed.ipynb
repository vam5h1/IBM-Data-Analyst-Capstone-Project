### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
# Step 3: Handling Missing Data

# Import necessary libraries
import pandas as pd

# Load dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Step 3.1: Check missing values per column
missing_values = df.isnull().sum().sort_values(ascending=False)
print("🔎 Top Columns with Missing Values:\n")
print(missing_values[missing_values > 0].head(20))

# Step 3.2: Define critical columns to retain
important_cols = ['Employment', 'JobSat', 'RemoteWork', 'YearsCodePro', 'LanguageHaveWorkedWith', 'Country']

# Drop rows with missing values in critical columns
df_cleaned = df.dropna(subset=important_cols)

# Confirm shape after dropping
print(f"\n✅ Data shape after dropping rows with missing values in {important_cols}: {df_cleaned.shape}")

# Step 3.3: Fill missing values in less critical columns (example: EdLevel, MainBranch)
df_cleaned['EdLevel'] = df_cleaned['EdLevel'].fillna('Unknown')
df_cleaned['MainBranch'] = df_cleaned['MainBranch'].fillna('Unknown')

# Final confirmation of cleaned data
print("\n✅ Final shape of cleaned dataset:", df_cleaned.shape)
print("✅ Sample data preview:")
print(df_cleaned[['Employment', 'JobSat', 'RemoteWork', 'YearsCodePro', 'LanguageHaveWorkedWith', 'Country']].head())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# Step 4: Analyze Key Columns

# 1. Employment Types Distribution
print("📊 Employment Type Distribution:\n")
print(df_cleaned['Employment'].value_counts(dropna=False), "\n")

# 2. Job Satisfaction Distribution
print("📊 Job Satisfaction Distribution:\n")
print(df_cleaned['JobSat'].value_counts(dropna=False), "\n")

# 3. Years of Professional Coding Experience Distribution
print("📊 Years of Professional Coding Experience Distribution:\n")
print(df_cleaned['YearsCodePro'].value_counts(dropna=False).sort_index(), "\n")

# Optional: Convert YearsCodePro to numeric for further analysis
def convert_experience(val):
    if val == 'Less than 1 year':
        return 0.5
    elif val == 'More than 50 years':
        return 51
    else:
        try:
            return float(val)
        except:
            return None

df_cleaned['YearsCodePro_Num'] = df_cleaned['YearsCodePro'].apply(convert_experience)

# Summary statistics for numeric experience
print("📈 Summary Statistics for Numeric Experience:\n")
print(df_cleaned['YearsCodePro_Num'].describe())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Step 5: KDE Plot for Job Satisfaction
plt.figure(figsize=(10, 6))
sns.kdeplot(df_cleaned['JobSat'].dropna(), shade=True, color='skyblue', bw_adjust=0.8)
plt.title('Job Satisfaction Distribution (KDE Plot)', fontsize=14)
plt.xlabel('Job Satisfaction Score (1 = Low, 10 = High)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.tight_layout()
plt.show()

# Optional: Pie Chart for Job Satisfaction levels
job_sat_counts = df_cleaned['JobSat'].value_counts().sort_index()
labels = job_sat_counts.index.astype(int).astype(str)
sizes = job_sat_counts.values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Blues", len(sizes)))
plt.title('Job Satisfaction (Proportional Distribution)', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
!pip install matplotlib-venn
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# STEP 1: Extract and count languages from both columns
def extract_lang_counts(df, column):
    lang_series = df[column].dropna().str.split(';')
    lang_list = [lang.strip() for sublist in lang_series for lang in sublist]
    return Counter(lang_list)

have_counts = extract_lang_counts(df_cleaned, 'LanguageHaveWorkedWith')
want_counts = extract_lang_counts(df_cleaned, 'LanguageWantToWorkWith')

# STEP 2: Create DataFrame for comparison
all_languages = list(set(have_counts) | set(want_counts))
lang_df = pd.DataFrame({
    'Language': all_languages,
    'HaveWorkedWith': [have_counts.get(lang, 0) for lang in all_languages],
    'WantToWorkWith': [want_counts.get(lang, 0) for lang in all_languages]
})

# Sort by total popularity
lang_df['Total'] = lang_df['HaveWorkedWith'] + lang_df['WantToWorkWith']
lang_df = lang_df.sort_values(by='Total', ascending=False).head(15)

# STEP 3: Plot grouped bar chart
plt.figure(figsize=(14, 6))
bar_width = 0.4
x = range(len(lang_df))

plt.bar(x, lang_df['HaveWorkedWith'], width=bar_width, label='Have Worked With', color='skyblue')
plt.bar([i + bar_width for i in x], lang_df['WantToWorkWith'], width=bar_width, label='Want to Work With', color='lightgreen')

plt.xlabel('Programming Languages')
plt.ylabel('Frequency')
plt.title('Top 15 Programming Languages: Have Worked With vs Want to Work With')
plt.xticks([i + bar_width / 2 for i in x], lang_df['Language'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
