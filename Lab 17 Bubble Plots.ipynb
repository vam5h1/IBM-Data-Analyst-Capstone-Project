### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data (assuming already downloaded and loaded as df)
# df = pd.read_csv("survey-data.csv")  # Already done in your case

# Inspect unique values for Age, SOPartFreq, and JobSat
print("Unique Age values:", df["Age"].dropna().unique())
print("Unique SOPartFreq values:", df["SOPartFreq"].dropna().unique())
print("JobSat unique values:", df["JobSat"].dropna().unique())

# Filter rows where Age, SOPartFreq, and JobSat are not null
df_filtered = df.dropna(subset=["Age", "SOPartFreq", "JobSat"])

# Since Age and SOPartFreq are categorical, map them to numeric values for plotting

# Map Age groups to numeric values (just ordinal)
age_order = [
    "Under 18 years old", "18-24 years old", "25-34 years old", "35-44 years old",
    "45-54 years old", "55-64 years old", "65 years or older", "Prefer not to say"
]
age_map = {age: i for i, age in enumerate(age_order)}
df_filtered["AgeNum"] = df_filtered["Age"].map(age_map)

# Map SOPartFreq to numeric values (custom order, approximate frequency)
freq_order = [
    "Never", "Less than once per year", "Once per year", "Multiple times per year",
    "Once per month", "Multiple times per month", "Once per week", "Multiple times per week",
    "Daily"
]
freq_map = {freq: i for i, freq in enumerate(freq_order)}
df_filtered["SOPartFreqNum"] = df_filtered["SOPartFreq"].map(freq_map)

# JobSat is categorical but ordinal; convert to numeric if necessary
# JobSat might be categorical labels; if so, map to numeric
job_sat_order = ["Very dissatisfied", "Somewhat dissatisfied", "Neither satisfied nor dissatisfied", "Somewhat satisfied", "Very satisfied"]
job_sat_map = {label: i+1 for i, label in enumerate(job_sat_order)}
df_filtered["JobSatNum"] = df_filtered["JobSat"].map(job_sat_map)

# Drop rows with any missing mappings
df_filtered = df_filtered.dropna(subset=["AgeNum", "SOPartFreqNum", "JobSatNum"])

# Plot bubble plot
plt.figure(figsize=(12, 8))

scatter = plt.scatter(
    df_filtered["AgeNum"],
    df_filtered["SOPartFreqNum"],
    s=df_filtered["JobSatNum"] * 100,  # Bubble size scaled by JobSatNum
    alpha=0.6,
    c=df_filtered["JobSatNum"],  # Color by JobSatNum for visual aid
    cmap="viridis",
    edgecolors="w",
    linewidth=0.5
)

# Customize axes ticks and labels
plt.xticks(ticks=list(age_map.values()), labels=list(age_map.keys()), rotation=45, ha="right")
plt.yticks(ticks=list(freq_map.values()), labels=list(freq_map.keys()))

plt.xlabel("Age Group")
plt.ylabel("Stack Overflow Participation Frequency")
plt.title("Bubble Plot: Age vs. SO Participation Frequency (Bubble Size = Job Satisfaction)")

# Add colorbar to indicate JobSat levels
cbar = plt.colorbar(scatter)
cbar.set_label("Job Satisfaction Level (1=Very dissatisfied, 5=Very satisfied)")

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming df is already loaded:
# df = pd.read_csv("survey-data.csv")

# Inspect unique Age values
print("Unique Age values:", df["Age"].dropna().unique())

# Map Age to numeric values (ordinal)
age_order = [
    "Under 18 years old", "18-24 years old", "25-34 years old", "35-44 years old",
    "45-54 years old", "55-64 years old", "65 years or older", "Prefer not to say"
]
age_map = {age: i for i, age in enumerate(age_order)}
df["AgeNum"] = df["Age"].map(age_map)

# Filter relevant data rows: no nulls in ConvertedCompYearly, JobSat, and AgeNum
df_filtered = df.dropna(subset=["ConvertedCompYearly", "JobSat", "AgeNum"])

# JobSat is categorical, map it to ordinal numeric scale for plotting
job_sat_order = ["Very dissatisfied", "Somewhat dissatisfied", "Neither satisfied nor dissatisfied", "Somewhat satisfied", "Very satisfied"]
job_sat_map = {label: i+1 for i, label in enumerate(job_sat_order)}
df_filtered["JobSatNum"] = df_filtered["JobSat"].map(job_sat_map)

# Drop rows with missing mappings
df_filtered = df_filtered.dropna(subset=["JobSatNum"])

# Scale bubble size (age): multiply to get visible sizes
bubble_size = (df_filtered["AgeNum"] + 1) * 100  # add 1 to avoid zero size

plt.figure(figsize=(12, 8))

scatter = plt.scatter(
    df_filtered["ConvertedCompYearly"],
    df_filtered["JobSatNum"],
    s=bubble_size,
    alpha=0.6,
    c=df_filtered["AgeNum"],
    cmap="plasma",
    edgecolors="w",
    linewidth=0.5
)

plt.xlabel("Yearly Compensation (USD)")
plt.ylabel("Job Satisfaction Level")
plt.title("Bubble Plot: Compensation vs Job Satisfaction (Bubble Size = Age)")

# Set y-axis ticks with labels
plt.yticks(ticks=list(range(1, 6)), labels=job_sat_order, rotation=45)

plt.colorbar(scatter, label="Age Group (Ordinal)")

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Assume df is loaded

# Step 1: Filter rows with non-null Age and LanguageHaveWorkedWith
df_filtered = df.dropna(subset=['Age', 'LanguageHaveWorkedWith'])

# Step 2: Explode the LanguageHaveWorkedWith column (semicolon-separated list)
df_filtered['LanguageHaveWorkedWith'] = df_filtered['LanguageHaveWorkedWith'].str.split(';')
df_exploded = df_filtered.explode('LanguageHaveWorkedWith')

# Step 3: Group by Age and Language, count occurrences
language_age_counts = df_exploded.groupby(['Age', 'LanguageHaveWorkedWith']).size().reset_index(name='Count')

# Optional: Filter to top N languages overall for clarity, e.g., top 10 languages
top_languages = df_exploded['LanguageHaveWorkedWith'].value_counts().head(10).index.tolist()
language_age_counts = language_age_counts[language_age_counts['LanguageHaveWorkedWith'].isin(top_languages)]

# Step 4: Pivot data for plotting
pivot_df = language_age_counts.pivot(index='Age', columns='LanguageHaveWorkedWith', values='Count').fillna(0)

# Sort Age groups in a logical order
age_order = [
    "Under 18 years old", "18-24 years old", "25-34 years old", "35-44 years old",
    "45-54 years old", "55-64 years old", "65 years or older", "Prefer not to say"
]
pivot_df = pivot_df.reindex(age_order)

# Step 5: Create the bubble plot
plt.figure(figsize=(14, 8))

for age_idx, age in enumerate(pivot_df.index):
    for lang_idx, lang in enumerate(pivot_df.columns):
        size = pivot_df.loc[age, lang]
        if size > 0:
            plt.scatter(lang_idx, age_idx, s=size*10, alpha=0.6, color='tab:blue', edgecolors='w')

plt.xticks(ticks=range(len(pivot_df.columns)), labels=pivot_df.columns, rotation=45, ha='right')
plt.yticks(ticks=range(len(pivot_df.index)), labels=pivot_df.index)
plt.xlabel('Programming Languages')
plt.ylabel('Age Group')
plt.title('Bubble Plot of Programming Languages by Age Group\n(Size = Frequency)')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is loaded

# Step 1: Clean and prepare data
# Drop rows with missing DatabaseWantToWorkWith or JobSat
df_filtered = df.dropna(subset=['DatabaseWantToWorkWith', 'JobSat'])

# Step 2: Explode if multiple databases listed (semicolon separated)
df_filtered['DatabaseWantToWorkWith'] = df_filtered['DatabaseWantToWorkWith'].str.split(';')
df_exploded = df_filtered.explode('DatabaseWantToWorkWith')

# Step 3: Group by database and job satisfaction, count respondents
grouped = df_exploded.groupby(['DatabaseWantToWorkWith', 'JobSat']).size().reset_index(name='Count')

# Step 4: Filter top 7 databases by total count for clarity
top_dbs = df_exploded['DatabaseWantToWorkWith'].value_counts().head(7).index
grouped = grouped[grouped['DatabaseWantToWorkWith'].isin(top_dbs)]

# Step 5: Pivot for plotting
pivot_df = grouped.pivot(index='JobSat', columns='DatabaseWantToWorkWith', values='Count').fillna(0)

# Sort JobSat categories if possible (e.g., from Very dissatisfied to Very satisfied)
job_sat_order = ['Very dissatisfied', 'Somewhat dissatisfied', 'Neither satisfied nor dissatisfied',
                 'Somewhat satisfied', 'Very satisfied']
pivot_df = pivot_df.reindex(job_sat_order)

# Step 6: Plot bubble plot
plt.figure(figsize=(14, 7))

for y_idx, job_sat in enumerate(pivot_df.index):
    for x_idx, db in enumerate(pivot_df.columns):
        size = pivot_df.loc[job_sat, db]
        if size > 0:
            plt.scatter(x_idx, y_idx, s=size*10, alpha=0.6, color='tab:green', edgecolors='w')

plt.xticks(ticks=range(len(pivot_df.columns)), labels=pivot_df.columns, rotation=45, ha='right')
plt.yticks(ticks=range(len(pivot_df.index)), labels=pivot_df.index)
plt.xlabel('Preferred Database to Work With Next Year')
plt.ylabel('Job Satisfaction')
plt.title('Bubble Plot of Preferred Databases vs Job Satisfaction\n(Size = Number of Respondents)')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Drop rows with missing DevType, Compensation or JobSat
df_filtered = df[['DevType', 'ConvertedCompYearly', 'JobSat']].dropna()

# Skip DevType entries that are floats (NaNs that somehow got through)
df_filtered = df_filtered[df_filtered['DevType'].apply(lambda x: isinstance(x, str))]

# Split multiple roles and explode
df_filtered['DevType'] = df_filtered['DevType'].str.split(';')
df_exploded = df_filtered.explode('DevType')

# Map JobSat to numeric scale
job_sat_map = {
    'Very dissatisfied': 1,
    'Somewhat dissatisfied': 2,
    'Neither satisfied nor dissatisfied': 3,
    'Somewhat satisfied': 4,
    'Very satisfied': 5
}
df_exploded['JobSatNum'] = df_exploded['JobSat'].map(job_sat_map)

# Drop rows with unmapped JobSat
df_exploded = df_exploded.dropna(subset=['JobSatNum'])

# Group by DevType
grouped = df_exploded.groupby('DevType').agg({
    'ConvertedCompYearly': 'median',
    'JobSatNum': 'mean',
    'DevType': 'count'
}).rename(columns={'DevType': 'Respondents'}).reset_index()

# Keep only DevTypes with valid numeric compensation
grouped = grouped[grouped['ConvertedCompYearly'].notna()]

# Select top 15 by response count for visibility
top_roles = grouped.sort_values('Respondents', ascending=False).head(15)

# Plot
plt.figure(figsize=(15, 8))
plt.scatter(
    x=top_roles['DevType'],
    y=top_roles['ConvertedCompYearly'],
    s=top_roles['JobSatNum'] * 300,  # bubble size
    alpha=0.6,
    c='skyblue',
    edgecolors='black'
)

plt.xticks(rotation=45, ha='right')
plt.xlabel("Developer Role")
plt.ylabel("Median Compensation (USD)")
plt.title("Compensation by Developer Role (Bubble Size = Avg Job Satisfaction)")
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Filter necessary columns and drop missing values
df_filtered = df[['Age', 'NEWCollabToolsHaveWorkedWith']].dropna()

# Ensure 'Age' is a string (handle edge cases)
df_filtered = df_filtered[df_filtered['Age'].apply(lambda x: isinstance(x, str))]

# Split collaboration tools (semicolon-separated) into separate rows
df_filtered['NEWCollabToolsHaveWorkedWith'] = df_filtered['NEWCollabToolsHaveWorkedWith'].str.split(';')
df_exploded = df_filtered.explode('NEWCollabToolsHaveWorkedWith')

# Count the frequency of each tool per age group
tool_counts = df_exploded.groupby(['Age', 'NEWCollabToolsHaveWorkedWith']).size().reset_index(name='Count')

# Optional: Filter for top 7 collaboration tools to keep the chart readable
top_tools = tool_counts['NEWCollabToolsHaveWorkedWith'].value_counts().nlargest(7).index
tool_counts = tool_counts[tool_counts['NEWCollabToolsHaveWorkedWith'].isin(top_tools)]

# Plot
plt.figure(figsize=(14, 8))
bubble = sns.scatterplot(
    data=tool_counts,
    x='Age',
    y='NEWCollabToolsHaveWorkedWith',
    size='Count',
    sizes=(100, 1000),
    hue='NEWCollabToolsHaveWorkedWith',
    palette='tab10',
    legend=False,
    alpha=0.6
)

plt.xticks(rotation=45, ha='right')
plt.title("Collaboration Tools Used by Age Group (Bubble Size = Usage Frequency)")
plt.xlabel("Age Group")
plt.ylabel("Collaboration Tool")
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Filter and clean necessary columns
df_filtered = df[['WebframeWantToWorkWith', 'JobSat']].dropna()

# Drop if JobSat is not string (guard clause)
df_filtered = df_filtered[df_filtered['JobSat'].apply(lambda x: isinstance(x, str))]

# Split multi-response frameworks into separate rows
df_filtered['WebframeWantToWorkWith'] = df_filtered['WebframeWantToWorkWith'].str.split(';')
df_exploded = df_filtered.explode('WebframeWantToWorkWith')

# Group and count respondents
framework_satisfaction = (
    df_exploded
    .groupby(['WebframeWantToWorkWith', 'JobSat'])
    .size()
    .reset_index(name='Count')
)

# Optional: Filter to top 7 most mentioned frameworks to declutter
top_frameworks = (
    framework_satisfaction['WebframeWantToWorkWith']
    .value_counts()
    .nlargest(7)
    .index
)
framework_satisfaction = framework_satisfaction[
    framework_satisfaction['WebframeWantToWorkWith'].isin(top_frameworks)
]

# Plotting
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=framework_satisfaction,
    x='WebframeWantToWorkWith',
    y='JobSat',
    size='Count',
    sizes=(100, 1200),
    hue='WebframeWantToWorkWith',
    alpha=0.6,
    legend=False
)

plt.title("Preferred Web Frameworks vs Job Satisfaction (Bubble Size = Respondent Count)")
plt.xlabel("Web Framework")
plt.ylabel("Job Satisfaction")
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
